```python
# Here we will check the percentage of nan values present in each feature

features_with_na=[features for features in taxi_df.columns if taxi_df[features].isnull().sum()>1]

if len(features_with_na)>0:
    for feature in features_with_na:
        print(feature, np.round(taxi_df[feature].isnull().mean(), 4),  ' % missing values')
else:
    print("No missing values")
```

```python
from sklearn.preprocessing import StandardScaler
features = ['fare','tips','tolls','extras','trip_seconds','trip_miles','pickup_latitude','dropoff_latitude']

x = taxi_df.loc[:, features].values
```

```python
y = taxi_df.loc[:,['trip_total']].values
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
principalComponents = pca.fit_transform(x)

principalDF = pd.DataFrame(data = principalComponents, columns = ['Principal Component 1' , 'Principal Component 2','Principal Component 3','Principal Component 4','Principal Component 5'])
```

```python
finalDf = pd.concat([principalDF , taxi_df[['trip_total']]], axis = 1)
finalDf
```

```python
per_var = np.round(pca.explained_variance_ratio_ * 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var) + 1)]
plt.bar(x=range(1, len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('percentange of explained variance')
plt.xlabel('principal component')
plt.title('screen plot')
plt.show()
```
